## Distribución de datos
Los datos de una variable pueden variar de muchas formas, es por ello que es importante estudiar como se distribuyen y para ello utilizamos medidas de `centralidad` y de `dispersión` así podemos observar que ocurre con dichos datos.

### Valores atípicos - Outliers

Un `valor atípico` o `outlier` es una observación que es numéricamente distante del resto de datos. Estos valores pueden ser indicativos de datos que pertenecen a una población diferente del resto de las muestras establecidas.

`¿Qué hacemos con esos valores atípicos?` La opción de ignorarlos a veces no es la mejor de las soluciones posibles si pretendemos conocer qué ha pasado con estos valores. Lo bien cierto es que distorsionan los resultados del análisis, por lo que hay que identificarlos y tratarlos de forma adecuada. `A veces se excluyen si son resultado de un error, pero otras veces son datos potencialmente interesantes en la detección de anomalías`.

Los valores atípicos pueden deberse a `errores` en la recolección de datos válidos que muestran un comportamiento diferente, pero reflejan la `aleatoriedad` de la variable en estudio. Es decir, valores que pueden haber aparecido como parte del proceso, aunque parezcan extraños. `Si los valores atípicos son parte del proceso, deben conservarse. En cambio, si ocurren por algún tipo de error (medida, codificación…), lo adecuado es su eliminación`.

Estos `outliers` tienen influencia en la `media`, pero no en la `mediana` o la `moda`. Para identificar dichos valores se pueden usar distintas técnicas, nosotros utilizaremos el rango intercuartilico ya que los datos de nuestras variables no se corresponden ni se asemejan a una `distribución normal`.

Para hacer una `estudio descriptivo (visual)` de la cantidad de valores atípicos usaremos un `grafico de caja` o `boxplot`. Para identificarlos matemáticamente utilizaremos el `Método de Tukey` que toma como referencia el `rango intercuartílico`.

## Visualización de datos tras eliminar `outliers`
En este caso, identificaremos cuales son los outliers para cada variable numérica y los eliminaremos para poder interpretar cómo varía la información tras su eliminación.

### Identificación y filtrado de valores `atípicos/outliers`

```python
# Total amount
Q1=df_taxi['total_amount'].quantile(0.25)
Q3=df_taxi['total_amount'].quantile(0.75)
IQR=Q3-Q1
outlier_sup=Q3+1.5*IQR
outlier_inf=Q1-1.5*IQR
print('Los límites para la variable "total_amount"')

print(f'Valor límite superior -> outlier_sup: {round(outlier_sup,3)}')
print(f'Valor límire inferior -> outlier_inf: {round(outlier_inf,3)} \n')

# Trip distance km
Q1_1=df_taxi['trip_distance_km'].quantile(0.25)
Q3_1=df_taxi['trip_distance_km'].quantile(0.75)
IQR_1=Q3_1-Q1_1
outlier_sup1=Q3_1+1.5*IQR_1
outlier_inf1=Q1_1-1.5*IQR_1
print('Los límites para la variable "trip_distance_km"')
print(f'Valor límite superior -> outlier_sup1: {round(outlier_sup1,3)}')
print(f'Valor límire inferior -> outlier_inf1: {round(outlier_inf1,3)} \n')

# Trip time h
Q1_2=df_taxi['trip_time_h'].quantile(0.25)
Q3_2=df_taxi['trip_time_h'].quantile(0.75)
IQR_2=Q3_2-Q1_2
outlier_sup2=Q3_2+1.5*IQR_2
outlier_inf2=Q1_2-1.5*IQR_2
print('Los límites para la variable "trip_time_h"')
print(f'Valor límite superior -> outlier_sup2: {round(outlier_sup2,3)}')
print(f'Valor límire inferior -> outlier_inf2: {round(outlier_inf2,3)} \n')

# Average speed kmh
Q1_3=df_taxi['average_speed_kmh'].quantile(0.25)
Q3_3=df_taxi['average_speed_kmh'].quantile(0.75)
IQR_3=Q3_3-Q1_3
outlier_sup3=Q3_3+1.5*IQR_3
outlier_inf3=Q1_3-1.5*IQR_3
print('Los límites para la variable "average_speed_kmh"')
print(f'Valor límite superior -> outlier_sup3: {round(outlier_sup3,3)}')
print(f'Valor límire inferior -> outlier_inf3: {round(outlier_inf3,3)} \n')
```

```python
df_taxi=df_taxi[(df_taxi['total_amount']<outlier_sup) & (df_taxi['trip_distance_km']<outlier_sup1) & (df_taxi['trip_time_h']<outlier_sup2) & (df_taxi['average_speed_kmh']<outlier_sup3)]
```

```python
df_taxi=reset_index(df_taxi,'index')
df_taxi.shape
```
